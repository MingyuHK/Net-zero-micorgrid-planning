```python
import pandas as pd

# 读取Excel文件
df = pd.read_excel('data/Loss.xlsx')

# 显示读取的数据（前几行）
print(df.values)
```

```python

import matplotlib.pyplot as plt# 创建一个新的图形
y = df.values
plt.figure(figsize=(8, 6))  # 设置图形的大小

# 绘制线图
plt.plot(y, marker='o', linestyle='-', color='b', label='Line 1')  # 指定标记、线型、颜色和图例标签

# 添加标题和标签
plt.title('Line Chart Example', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

# 添加网格线
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# 添加图例
plt.legend()

# 显示图形
plt.show()
```
